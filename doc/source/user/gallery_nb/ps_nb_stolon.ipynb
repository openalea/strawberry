```python
import cufflinks as cf
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)

from openalea.strawberry.import_mtgfile import import_mtgfile
from openalea.strawberry.analysis import extract_at_plant_scale
from openalea.strawberry.application.app_plant_scale import plot_plantscale
```

```python
Gariguette = import_mtgfile('Gariguette')
df = extract_at_plant_scale(Gariguette)

res = df.groupby('date').mean()
res.iplot(kind = "line", 
              mode='lines+markers', 
              y='nb_stolons',
              xTitle="Dates",
              yTitle='nb_stolons',
              asFigure=True
             )
```
